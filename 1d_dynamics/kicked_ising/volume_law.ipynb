```python
from funcs import *
import pickle
from matplotlib import pyplot as plt
import matplotlib as mpl
import matplotlib.cm as mplcm
import numpy as np
import matplotlib.colors as colors

print(mpl.get_cachedir())
```

```python
import matplotlib.pyplot as plt
plt.rcParams['mathtext.fontset'] = 'custom'
plt.rcParams['mathtext.rm'] = 'Arial'
plt.rcParams['mathtext.bf'] = 'Arial:bold'
plt.rcParams['mathtext.it'] = 'Arial:italic'
plt.rcParams['mathtext.sf'] = 'Arial'
plt.rcParams['mathtext.tt'] = 'Arial'
plt.rcParams['font.family'] = 'Arial'
# plt.title(r'ABC123 vs $\mathrm{ABC123}^{123}$ vs $J$')
```

```python
n = 14
J = 0.7
g = 0.5
h = 0.5
depth = 20
# Load the data
with open(f'./results/volume_law/exact_n={n}_J={J}_g={g}_h={h}.pkl', 'rb') as f:
    exact = pickle.load(f)


NUM_COLORS = 20
cm = plt.get_cmap('turbo')
cNorm  = colors.Normalize(vmin=0, vmax=NUM_COLORS-1)
scalarMap = mplcm.ScalarMappable(norm=cNorm, cmap=cm)
# Plot the data, format: {t: l-s list}
fig, ax = plt.subplots(figsize=(8, 6))
# ax.set_prop_cycle(color=[scalarMap.to_rgba(i) for i in range(NUM_COLORS)])
ax.set_prop_cycle(color=[cm(1.*i/NUM_COLORS) for i in range(NUM_COLORS)])
plt.rcParams['mathtext.fontset'] = 'custom'
plt.rcParams['mathtext.rm'] = 'Arial'
plt.rcParams['mathtext.bf'] = 'Arial:bold'
plt.rcParams['mathtext.it'] = 'Arial:italic'
plt.rcParams['mathtext.sf'] = 'Arial'
plt.rcParams['mathtext.tt'] = 'Arial'
plt.rcParams['font.family'] = 'Arial'

for t, data in exact.items():
    ax.plot(data, label=f'$t={t}$', marker='o', linestyle='-', markersize=7, linewidth=1.2)
ax.set_xlabel(r'subsystem size $L_B$', fontsize=14)
ax.set_ylabel(r'$S_B(L_B)$', fontsize=14)
# ax.set_title(f'Entanglement entropy spatial scaling for $n={n}$, $J={J}$, $g=h={h}$')
ax.text(0.2, 0.6, f'${n}$ qubits, $J={J}$, $g=h={h}$', horizontalalignment='center', verticalalignment='center', transform=ax.transAxes, fontsize=12)
ax.minorticks_on()
# from matplotlib.ticker import AutoMinorLocator
# ax.xaxis.set_minor_locator(AutoMinorLocator())
# ax.yaxis.set_minor_locator(AutoMinorLocator())
ax.tick_params(axis='both',          # Apply changes to both x and y axes
               direction='in',       # Set ticks to face inward
               length=6,             # Tick length in points
               width=1.5,              # Tick width in points
            #    colors='b',           # Color of the ticks
            #    grid_color='r',       # Color of the gridlines
               grid_alpha=0.5,       # Transparency of gridlines
               right=True,           
               top=True,
               labelsize=12
)
ax.tick_params(axis='both', which='minor', length=3, width=1.3, color='black', direction='in', right=True)
# set axis thickness
for axis in ['top','bottom','left','right']:
  ax.spines[axis].set_linewidth(1.5)
ax.set_xlim(0, n//2-1)
ax.set_ylim(0, 6.8)
# change legend position
ax.legend(ncol=4, frameon=False, loc='upper left', bbox_to_anchor=(0.02, 1), fontsize=10)
plt.savefig(f'./results/volume_law/less_chaotic.pdf', dpi=300, bbox_inches='tight')
plt.show()
```
